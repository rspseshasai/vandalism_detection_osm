## Hyper Classifier model for Changeset data

#### 1. Imports and Setup

```python
# Import necessary libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the paths (update these as needed)
AGGREGATED_FEATURES_PATH = VISUALIZATION_DATA_PATH['hyper_classifier_features_sample_path']
EVALUATION_RESULTS_PATH = VISUALIZATION_DATA_PATH['evaluation_results_hyper_classifier']
CONFUSION_MATRIX_PATH = VISUALIZATION_DATA_PATH['confusion_matrix_hyper_classifier']

# Helper function to set consistent plot styles
def set_plot_style():
    sns.set(style="whitegrid")
    plt.rcParams.update({"font.size": 12})
```

#### 2. Visualize Aggregated Features

```python
# Load aggregated features
aggregated_features = pd.read_parquet(AGGREGATED_FEATURES_PATH)

# Display the first few rows
print("Aggregated Hyper Classifier Features Sample:")
aggregated_features.head()
```

```python
# Plot distributions of key features
set_plot_style()

# Number of contributions
plt.figure(figsize=(8, 5))
sns.histplot(aggregated_features['num_contributions'], bins=30, kde=True, color='blue')
plt.title('Distribution of Number of Contributions per Changeset')
plt.xlabel('Number of Contributions')
plt.ylabel('Frequency')
plt.show()

# Mean prediction
plt.figure(figsize=(8, 5))
sns.histplot(aggregated_features['mean_prediction'], bins=30, kde=True, color='green')
plt.title('Distribution of Mean Predictions')
plt.xlabel('Mean Prediction')
plt.ylabel('Frequency')
plt.show()

# Standard deviation of predictions
plt.figure(figsize=(8, 5))
sns.histplot(aggregated_features['std_prediction'], bins=30, kde=True, color='orange')
plt.title('Distribution of Standard Deviation of Predictions')
plt.xlabel('Standard Deviation')
plt.ylabel('Frequency')
plt.show()
```

#### 3. Visualize Evaluation Results

```python
# Load evaluation results
evaluation_results = pd.read_parquet(EVALUATION_RESULTS_PATH)

# Display the first few rows
print("Evaluation Results Sample:")
evaluation_results.head()
```

```python
plot_confusion_matrix('hyper_classifier')
```

```python
# Calculate and display key metrics from evaluation results
accuracy = (evaluation_results['y_true'] == evaluation_results['y_pred_hyper_classifier']).mean()
precision = evaluation_results.query("y_pred_hyper_classifier == 1 & y_true == 1").shape[0] / evaluation_results.query("y_pred_hyper_classifier == 1").shape[0]
recall = evaluation_results.query("y_pred_hyper_classifier == 1 & y_true == 1").shape[0] / evaluation_results.query("y_true == 1").shape[0]
f1_score = 2 * (precision * recall) / (precision + recall)

print("Evaluation Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1_score:.4f}")
```

#### 4. Visualize ROC and PR Curves

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Calculate ROC Curve
fpr, tpr, _ = roc_curve(evaluation_results['y_true'], evaluation_results['y_prob_hyper_classifier'])

# Plot ROC Curve
plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(evaluation_results['y_true'], evaluation_results['y_prob_hyper_classifier']):.4f}")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

# Calculate Precision-Recall Curve
precision, recall, _ = precision_recall_curve(evaluation_results['y_true'], evaluation_results['y_prob_hyper_classifier'])

# Plot Precision-Recall Curve
plt.figure(figsize=(8, 5))
plt.plot(recall, precision, label=f"Avg Precision = {average_precision_score(evaluation_results['y_true'], evaluation_results['y_prob_hyper_classifier']):.4f}")
plt.title("Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend(loc="lower left")
plt.grid()
plt.show()
```

## Meta Classifier model for Changeset Data

#### 1. Import Necessary Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import (
    roc_curve, auc, confusion_matrix, precision_recall_curve,
    classification_report
)
```

#### 2. Load Results

```python
# Load the results
results_df = pd.read_csv(VISUALIZATION_DATA_PATH['evaluation_results_meta_classifier'])
```

#### 3. Print Performance Metrics

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score

def calculate_metrics(y_true, y_pred, y_prob):
    metrics = {}
    metrics['Accuracy'] = accuracy_score(y_true, y_pred)
    metrics['Precision'] = precision_score(y_true, y_pred, zero_division=0)
    metrics['Recall'] = recall_score(y_true, y_pred, zero_division=0)
    metrics['F1-Score'] = f1_score(y_true, y_pred, zero_division=0)
    metrics['AUC-ROC'] = roc_auc_score(y_true, y_prob)
    return metrics

# Calculate metrics for each model
metrics_main = calculate_metrics(
    results_df['y_true'],
    results_df['y_pred_main_model'],
    results_df['y_prob_main_model']
)

metrics_hyper = calculate_metrics(
    results_df['y_true'],
    results_df['y_pred_hyper_classifier'],
    results_df['y_prob_hyper_classifier']
)

metrics_meta = calculate_metrics(
    results_df['y_true'],
    results_df['y_pred_meta_classifier'],
    results_df['y_prob_meta_classifier']
)

# Create a DataFrame for metrics
metrics_df = pd.DataFrame({
    'Main Model': metrics_main,
    'Hyper-Classifier Model': metrics_hyper,
    'Meta-Classifier Model': metrics_meta
})

# Transpose for better display
metrics_df = metrics_df.transpose()

# Display the metrics table
print("Performance Metrics:")
display(metrics_df.style.format("{:.4f}"))
```

#### 4. Plot ROC Curves

```python
# Plot ROC Curves
plt.figure(figsize=(10, 8))

# Main Model
fpr_main, tpr_main, _ = roc_curve(results_df['y_true'], results_df['y_prob_main_model'])
roc_auc_main = auc(fpr_main, tpr_main)
plt.plot(fpr_main, tpr_main, label=f'Main Model (AUC = {roc_auc_main:.2f})')

# Hyper-Classifier Model
fpr_hyper, tpr_hyper, _ = roc_curve(results_df['y_true'], results_df['y_prob_hyper_classifier'])
roc_auc_hyper = auc(fpr_hyper, tpr_hyper)
plt.plot(fpr_hyper, tpr_hyper, label=f'Hyper-Classifier Model (AUC = {roc_auc_hyper:.2f})')

# Meta-Classifier Model
fpr_meta, tpr_meta, _ = roc_curve(results_df['y_true'], results_df['y_prob_meta_classifier'])
roc_auc_meta = auc(fpr_meta, tpr_meta)
plt.plot(fpr_meta, tpr_meta, label=f'Meta-Classifier Model (AUC = {roc_auc_meta:.2f})')

# Plot settings
plt.plot([0, 1], [0, 1], 'k--')  # Diagonal line
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate', fontsize=14)
plt.ylabel('True Positive Rate', fontsize=14)
plt.title('Receiver Operating Characteristic (ROC) Curves', fontsize=16)
plt.legend(loc='lower right', fontsize=12)
plt.grid(True)
plt.show()
```

#### 5. Plot Confusion Matrices

```python
from sklearn.metrics import ConfusionMatrixDisplay

models = ['Main Model', 'Hyper-Classifier Model', 'Meta-Classifier Model']
predictions = [
    ('Main Model', results_df['y_pred_main_model']),
    ('Hyper-Classifier Model', results_df['y_pred_hyper_classifier']),
    ('Meta-Classifier Model', results_df['y_pred_meta_classifier'])
]

for model_name, y_pred in predictions:
    cm = confusion_matrix(results_df['y_true'], y_pred)
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Non-Vandalism', 'Vandalism'])
    disp.plot(cmap=plt.cm.Blues)
    plt.title(f'Confusion Matrix: {model_name}', fontsize=16)
    plt.xlabel('Predicted Label', fontsize=14)
    plt.ylabel('True Label', fontsize=14)
    plt.show()
```

#### 6. Plot Precision-Recall Curves

```python
# Plot Precision-Recall Curves
plt.figure(figsize=(10, 8))

# Main Model
precision_main, recall_main, _ = precision_recall_curve(results_df['y_true'], results_df['y_prob_main_model'])
plt.plot(recall_main, precision_main, label='Main Model')

# Hyper-Classifier Model
precision_hyper, recall_hyper, _ = precision_recall_curve(results_df['y_true'], results_df['y_prob_hyper_classifier'])
plt.plot(recall_hyper, precision_hyper, label='Hyper-Classifier Model')

# Meta-Classifier Model
precision_meta, recall_meta, _ = precision_recall_curve(results_df['y_true'], results_df['y_prob_meta_classifier'])
plt.plot(recall_meta, precision_meta, label='Meta-Classifier Model')

# Plot settings
plt.xlabel('Recall', fontsize=14)
plt.ylabel('Precision', fontsize=14)
plt.title('Precision-Recall Curves', fontsize=16)
plt.legend(loc='lower left', fontsize=12)
plt.grid(True)
plt.show()
```

#### 7. Classification Reports

```python
# Classification Reports
for model_name, y_pred in predictions:
    print(f"Classification Report for {model_name}:")
    report = classification_report(results_df['y_true'], y_pred, target_names=['Non-Vandalism', 'Vandalism'], zero_division=0)
    print(report)
    print("-" * 80)
```
