## Hyper Classifier model for Changeset data

#### 1. Imports and Setup

```python
from config import HYPER_VISUALIZATION_DIR
# Import necessary libraries
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Define the paths (update these as needed)
AGGREGATED_FEATURES_PATH = os.path.join(HYPER_VISUALIZATION_DIR, 'aggregated_features_sample.parquet')
EVALUATION_RESULTS_PATH = os.path.join(HYPER_VISUALIZATION_DIR, 'evaluation_results.parquet')
CONFUSION_MATRIX_PATH = os.path.join(HYPER_VISUALIZATION_DIR, 'confusion_matrix.npy')

# Helper function to set consistent plot styles
def set_plot_style():
    sns.set(style="whitegrid")
    plt.rcParams.update({"font.size": 12})
```

#### 2. Visualize Aggregated Features

```python
# Load aggregated features
aggregated_features = pd.read_parquet(AGGREGATED_FEATURES_PATH)

# Display the first few rows
print("Aggregated Features Sample:")
aggregated_features.head()
```

```python
# Plot distributions of key features
set_plot_style()

# Number of contributions
plt.figure(figsize=(8, 5))
sns.histplot(aggregated_features['num_contributions'], bins=30, kde=True, color='blue')
plt.title('Distribution of Number of Contributions per Changeset')
plt.xlabel('Number of Contributions')
plt.ylabel('Frequency')
plt.show()

# Mean prediction
plt.figure(figsize=(8, 5))
sns.histplot(aggregated_features['mean_prediction'], bins=30, kde=True, color='green')
plt.title('Distribution of Mean Predictions')
plt.xlabel('Mean Prediction')
plt.ylabel('Frequency')
plt.show()

# Standard deviation of predictions
plt.figure(figsize=(8, 5))
sns.histplot(aggregated_features['std_prediction'], bins=30, kde=True, color='orange')
plt.title('Distribution of Standard Deviation of Predictions')
plt.xlabel('Standard Deviation')
plt.ylabel('Frequency')
plt.show()
```

#### 3. Visualize Evaluation Results

```python
# Load evaluation results
evaluation_results = pd.read_parquet(EVALUATION_RESULTS_PATH)

# Display the first few rows
print("Evaluation Results Sample:")
evaluation_results.head()
```

```python
# Load confusion matrix
confusion_matrix = np.load(CONFUSION_MATRIX_PATH)

# Plot confusion matrix
set_plot_style()

plt.figure(figsize=(6, 5))
sns.heatmap(
    confusion_matrix, annot=True, fmt='d', cmap='Blues',
    xticklabels=['Non-Vandalism', 'Vandalism'],
    yticklabels=['Non-Vandalism', 'Vandalism']
)
plt.title('Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

```python
# Calculate and display key metrics from evaluation results
accuracy = (evaluation_results['y_true'] == evaluation_results['y_pred']).mean()
precision = evaluation_results.query("y_pred == 1 & y_true == 1").shape[0] / evaluation_results.query("y_pred == 1").shape[0]
recall = evaluation_results.query("y_pred == 1 & y_true == 1").shape[0] / evaluation_results.query("y_true == 1").shape[0]
f1_score = 2 * (precision * recall) / (precision + recall)

print("Evaluation Metrics:")
print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1 Score: {f1_score:.4f}")
```

#### 4. Visualize ROC and PR Curves

```python
from sklearn.metrics import roc_curve, roc_auc_score

# Calculate ROC Curve
fpr, tpr, _ = roc_curve(evaluation_results['y_true'], evaluation_results['y_prob'])

# Plot ROC Curve
plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, label=f"AUC = {roc_auc_score(evaluation_results['y_true'], evaluation_results['y_prob']):.4f}")
plt.title("Receiver Operating Characteristic (ROC) Curve")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend(loc="lower right")
plt.grid()
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve, average_precision_score

# Calculate Precision-Recall Curve
precision, recall, _ = precision_recall_curve(evaluation_results['y_true'], evaluation_results['y_prob'])

# Plot Precision-Recall Curve
plt.figure(figsize=(8, 5))
plt.plot(recall, precision, label=f"Avg Precision = {average_precision_score(evaluation_results['y_true'], evaluation_results['y_prob']):.4f}")
plt.title("Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend(loc="lower left")
plt.grid()
plt.show()
```
